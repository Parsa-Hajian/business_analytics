# Simple Decision Tree

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load and prepare
df = pd.read_csv("bank-additional-full.csv", sep=";")
df["y"] = (df["y"] == "yes").astype(int)
features = ["age", "campaign", "previous", "contact", "month"]
X = pd.get_dummies(df[features], drop_first=True)
y = df["y"]

# Fit small tree
tree = DecisionTreeClassifier(max_depth=3, min_samples_leaf=200)
tree.fit(X, y)

# Plot
plt.figure(figsize=(14, 8))
plot_tree(tree, feature_names=X.columns, class_names=["no", "yes"], filled=True, rounded=True)
plt.title("Decision Tree: Campaign Subscription Segmentation")
plt.tight_layout()
plt.show()
```
