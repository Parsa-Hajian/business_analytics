### 3.2 Understanding Our Dataset

#### Goal:
Before jumping into building predictive models, it's critical to understand the dataset. Models built on misunderstood data can lead to misleading or irrelevant results. By examining the structure, distribution, and relationships within the data, we can ensure that the analysis aligns with business goals.

#### Key Components:
We break down the dataset into the following functional groups:

1. **Target and Base Rate**:
   - **Target Variable (y)**: Subscription outcome (yes/no).
   - **Base Rate**: Approximately 11% of the clients subscribe to a term deposit. A low base rate means that simply predicting "no" would achieve high accuracy, making **accuracy** a misleading metric.
   - **Importance**: Understanding the base rate helps us recognize how segmentation and KPIs like **Response Lift** can improve upon a naive model.

2. **Customer Profile (Who Are We Talking To?)**:
   - Variables like **age**, **job**, **marital status**, and **education** give us insight into who our customers are.
   - Demographics, financial attributes (like **loan** and **default** status), and behavior are critical to segmenting our audience.

3. **Campaign Interaction (How and When We Contacted)**:
   - **Contact Method**: Cellular vs. telephone, which influences response rates.
   - **Timing**: The **month** and **day of the week** can significantly impact the campaignâ€™s success.
   - **Contact Frequency**: Too many contacts can result in **contact fatigue**; understanding the optimal number of attempts is essential.

4. **Prior Engagement (What Happened Before)**:
   - Past interactions, such as **previous contact** or **previous campaign outcomes**, are often the strongest predictors of future behavior.

5. **Economic Context**:
   - Variables like **employment variation rate**, **consumer confidence index**, and **interest rates** show how external conditions affect customer behavior and decision-making.

#### Strategy from EDA:
- Segment the data based on demographic and economic attributes, campaign interactions, and past behavior.
- Analyze the relationships between **features** (age, job, marital status) and the **target variable** to identify key drivers of success.
- Visualize how economic conditions correlate with campaign success to adjust strategies.

The next steps involve applying EDA to break down each of these components. We will look at the **target variable**, explore the **customer profile**, examine **campaign interactions**, and analyze **economic conditions**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("bank-additional-full.csv", sep=";")

# Step 1: Analyze the Target Variable (y) - Subscription Outcome
plt.figure(figsize=(6, 4))
sns.countplot(x='y', data=df)
plt.title('Subscription Outcome Distribution')
plt.xlabel('Subscription (Yes/No)')
plt.ylabel('Count')
plt.show()

# Check the base rate (subscription rate)
base_rate = df['y'].value_counts(normalize=True)
print(f"Base Rate (subscription rates):\n{base_rate}")
```

```python
# Step 2.1: Age Distribution by Subscription Outcome (Boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='y', y='age', data=df)
plt.title('Age Distribution by Subscription Outcome')
plt.xlabel('Subscription (Yes/No)')
plt.ylabel('Age')
plt.show()
```

```python
# Step 2.2: Subscription Rate by Job Type (Countplot)
plt.figure(figsize=(12, 8))
sns.countplot(x='job', hue='y', data=df)
plt.title('Subscription Rate by Job Type')
plt.xlabel('Job Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 2.3: Subscription Rate by Marital Status
plt.figure(figsize=(8, 6))
sns.countplot(x='marital', hue='y', data=df)
plt.title('Subscription Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
# Step 3.1: Subscription Rate by Contact Channel
plt.figure(figsize=(8, 6))
sns.countplot(x='contact', hue='y', data=df)
plt.title('Subscription Rate by Contact Channel')
plt.xlabel('Contact Method')
plt.ylabel('Count')
plt.show()
```

```python
# Step 3.2: Subscription Rate by Month
plt.figure(figsize=(10, 6))
sns.countplot(x='month', hue='y', data=df)
plt.title('Subscription Rate by Month of Contact')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 3.3: Subscription Rate by Contact Frequency
plt.figure(figsize=(10, 6))
sns.countplot(x='campaign', hue='y', data=df)
plt.title('Subscription Rate by Contact Frequency')
plt.xlabel('Number of Contacts')
plt.ylabel('Count')
plt.show()
```

```python
# Step 4: Economic Context - Employment Variation Rate and Subscription Outcome
plt.figure(figsize=(10, 6))
sns.boxplot(x='y', y='emp.var.rate', data=df)
plt.title('Employment Variation Rate by Subscription Outcome')
plt.xlabel('Subscription (Yes/No)')
plt.ylabel('Employment Variation Rate')
plt.show()
```
